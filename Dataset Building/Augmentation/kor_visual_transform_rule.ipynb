## Load Dataset

## Apply transformation function

```python
# Transformation function
def apply_transformation(value, rules):
    """
    Apply character-based transformation to a given value using predefined rules.

    Args:
        value (str or any): The input value to transform. If NaN, it remains unchanged.
        rules (list of tuples): A list of (original, replacement) character mappings.

    Returns:
        str or any: Transformed value if applicable, otherwise the original value.
    """
    if pd.isna(value):
        return value
    transformed = ""
    for char in str(value):
        for original, replacement in rules:
            if char == original:
                char = replacement
                break
        transformed += char
    return transformed

# Apply transformation to JSON data
transformed_data = [
    {"original": item["value"], "transformed": apply_transformation(item["value"], transformation_rules)}
    for item in json_data
]
```

## Save to JSON

```python
# Save transformed data as a JSON file
with open(output_json_path, "w", encoding="utf-8") as json_file:
    json.dump(transformed_data, json_file, ensure_ascii=False, indent=4)
print(f"Save transformed data as a JSON file: {output_json_path}")
```
